```python
import os
import shutil

def create_or_delete_folder(folder_name):
    if not os.path.exists(folder_name):
        os.mkdir(folder_name)
    else:
        shutil.rmtree(folder_name)
        os.mkdir(folder_name)

create_or_delete_folder("my_folder")

import matplotlib.pyplot as plt
import numpy as np

# Створення послідовності значень x від -10 до 10
x = np.linspace(-10, 10, 100)  # 100 точок між -10 і 10

# Обчислення відповідних значень y для параболи
y = x**2

# Створення графіка
plt.plot(x, y)

# Додавання підписів для осей та заголовку графіку
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік параболи: y = x^2')

# Виведення графіку
plt.show()
```
